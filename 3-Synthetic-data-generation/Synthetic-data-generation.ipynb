###	Draw 2 clusters 

```python
X, y = make_blobs(n_samples=200, centers=[[1.2,4],[4.2,1]], n_features=2)
df = DataFrame(dict(x=X[:,0], y=X[:,1], label=y))
colors = {0:'orange', 1:'green'}
fig, ax = pyplot.subplots()
grouped = df.groupby('label')
for key, group in grouped:
    group.plot(ax=ax, kind='scatter', x='x', y='y', label=key, color=colors[key])
pyplot.legend(["Tigers","Lions"])
pyplot.show()
```

### Draw 5 classes creating a dataset

```python
data, labels = make_circles(n_samples=100, shuffle=True, noise=0.05, random_state=42)
data2, labels2 = make_blobs(n_samples=100, cluster_std = 0.5,centers=[[3, 4], [5, 3], [4.5, 6]],random_state=1)
centers=[[3, 4], [5, 3], [4.5, 6]]

for i in range(len(centers)-1, -1, -1):
    labels2[labels2==0+i] = i+2

labels = np.concatenate([labels, labels2])
data = data * [1.2, 1.8] + [3, 4]

data = np.concatenate([data, data2], axis=0)
fig, ax = pyplot.subplots()

colours = ["orange", "blue", "magenta", "yellow", "green"]
label_name = ["Tiger", "Lion", "Penguin", "Dolphin", "Python"]

for label in range(0, len(centers)+2):
    ax.scatter(data[labels==label, 0], data[labels==label, 1], c=colours[label], s=40, label=label_name[label])

pyplot.title('dataset')
pyplot.legend(loc='upper right')
pyplot.show()
```

### Using sklearn Fetch the Olivetti faces dataset and visualize the faces

```python
faces = fetch_olivetti_faces()
fig = pyplot.figure(figsize=(8, 8))
fig.subplots_adjust(left=0, right=1, bottom=0, top=1)
for i in range(64):
    ax = fig.add_subplot(8, 8, i + 1, xticks=[], yticks=[])
    ax.imshow(faces.images[i], cmap=pyplot.cm.bone, interpolation='nearest')
    ax.text(0, 7, str(faces.target[i]))
```
